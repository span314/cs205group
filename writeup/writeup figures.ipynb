```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


%matplotlib inline
```

```python
apsp_results = pd.read_csv(open('../apsp_results.csv'))
apsp_results
```

```python
plt.plot(np.exp2(apsp_results.scale), apsp_results.regtrop, marker='o', label='Tropical Semiring (row-major)')
plt.plot(np.exp2(apsp_results.scale), apsp_results.quadtrop, marker='^', label='Tropical Semiring (quadtree)')
plt.plot(np.exp2(apsp_results.scale)[:-1], apsp_results.floyd[:-1], marker='v', label='Floyd-Warshall (sequential)')
plt.xscale('log', basex=2)
plt.yscale('log')
plt.ylabel('Execution time (ms)')
plt.xlabel('# nodes in graph')
plt.legend(loc='best', fontsize=10)
plt.title('APSP Performance Results')
plt.show()
```

```python
bfs_results = pd.read_csv(open('../bfs_results.csv'))
bfs_results
```

```python
plt.plot(np.exp2(bfs_results.scale), bfs_results.edgelist_bfs, marker='o', label='Edgelist BFS')
plt.plot(np.exp2(bfs_results.scale), bfs_results.edgelist_bfs_parallel, marker='^', label='Edgelist BFS Parallel')
plt.xscale('log', basex=2)
plt.yscale('log')
plt.ylabel('Execution time (ms)')
plt.xlabel('# nodes in graph')
plt.legend(loc='best', fontsize=10)
plt.title('BFS Performance Results')
plt.show()
```
